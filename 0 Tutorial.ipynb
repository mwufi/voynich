# Tutorial

* Load Voynich data
* do some stuff

```python
# here's the public interface
from vdata import Voynich, Folio
```

```python
a = Voynich()
```

```python
a.data.head()
```

The main data structure I'm using is a Pandas dataframe. (yes, i know, this is not a database and not a text file)
* you can filter by folio, line number, or the length of the line!
* conveniently, the Voynich class contains two useful methods

```python
# retrieve the lines on a page
a.getLines('f10v')
```

```python
# retrieve the words on a page
a.getWords('f10v')
```

```python
# an ordered list of the folios
# let's print only the first 10 for convenience
a.folios[:10]
```

```python
# we can also pick out a single folio
f = Folio('f10v')
```
